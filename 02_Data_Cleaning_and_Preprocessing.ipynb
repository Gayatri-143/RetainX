```python
import pandas as pd
```

```python
df = pd.read_csv(r'cleaned_retail.csv')
```

```python
df_clean = df.copy()
```

```python
cancelled = df_clean[df_clean['Invoice'].astype(str).str.startswith('C')]
```

```python
cancelled.shape
```

```python
df_clean = df_clean[~df_clean['Invoice'].astype(str).str.startswith('C')]
```

```python
# Removed cancelled invoices (Invoice starting with 'C') to ensure only successful transactions are analyzed
```

```python
df_clean['Customer ID'].isnull().sum()
```

```python
df_clean = df_clean.dropna(subset=['Customer ID'])
```

# Removing negative quantities

```python
df_clean[df_clean['Quantity'] < 0].shape
```

```python
df_clean = df_clean[df_clean['Quantity'] > 0]
```

```python
# Removed transactions with negative quantities to ensure only valid purchases are analyzed.
```

# Converting InvoiceDate to Datetime

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
```

```python
df_clean.info()
```

```python
# creating revenue column
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['Price']
```
