# Revenue Column

```python
import pandas as pd

df_clean = pd.read_csv("cleaned_online_retail.csv")
```

```python
# Reveue = Quantity X Price
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['Price']
```

```python
df_clean[['Quantity','Price','Revenue']].head()
# Created a Revenue feature by multiplying Quantity with UnitPrice to represent total transaction value.
```

# Extract Order Month

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
```

```python
df_clean['OrderMonth'] = df_clean['InvoiceDate'].dt.month
```

```python
df_clean[['InvoiceDate','OrderMonth']].head()
# Extracted OrderMonth from InvoiceDate to analyze monthly sales trends.
```
