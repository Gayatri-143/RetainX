# Revenue Column

```python
import pandas as pd

df_clean = pd.read_csv("cleaned_online_retail.csv")
```

```python
# Reveue = Quantity X Price
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['Price']
```

```python
df_clean[['Quantity','Price','Revenue']].head()
# Created a Revenue feature by multiplying Quantity with UnitPrice to represent total transaction value.
```

# Extract Order Month

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
```

```python
df_clean['OrderMonth'] = df_clean['InvoiceDate'].dt.month
```

```python
df_clean[['InvoiceDate','OrderMonth']].head()
# Extracted OrderMonth from InvoiceDate to analyze monthly sales trends.
```

# Extract Order Year

```python
df_clean['OrderYear'] = df_clean['InvoiceDate'].dt.year
```

```python
df_clean[['InvoiceDate','OrderYear']].head(20)
# Extracted OrderYear to enable yearly sales comparisons.
```

# Creating Customer Tenure

```python
# Customer Tenure : How long a customer has been active since their first purchase.
```

```python
# SQL fromula
#Tenure = Current Order Date - Customer First Purchase Date
```

```python
# first purchase per customer
first_purchase = df_clean.groupby('Customer ID')['InvoiceDate'].min()
```

```python
# first purchase to dataset
df_clean['FirstPurchaseDate'] = df_clean['Customer ID'].map(first_purchase)
```

```python
# Calculate tenure
df_clean['CustomerTenureDays'] = (df_clean['InvoiceDate'] - df_clean['FirstPurchaseDate']).dt.days
```

```python
df_clean[['Customer ID','InvoiceDate','FirstPurchaseDate','CustomerTenureDays']].head()
```

```python
# tenure is calculated as the difference between the transaction date and the customer's first purchase date, 
# this helps analyze customer loyalty and behaviour
```
